```python
corpus = """
Hello This is Aryan Singh learning NLP and Tokenization.
I want to bea data driven engineer with hands not only on data analytics and engineering field! but also in artificial intelligence.
"""
```

```python
from nltk.tokenize import sent_tokenize
```

```python
import nltk
# Check if 'punkt' tokenizer data is available
try:
    nltk.data.find('tokenizers/punkt/english.pickle')
# If 'punkt' is not found (LookupError)
except LookupError:
    print("NLTK 'punkt' tokenizer data not found. Downloading...")
    # Download the 'punkt' tokenizer data
    nltk.download('punkt')
    print("Download complete.")

# Download the 'punkt_tab' resource as suggested by the error
try:
    nltk.data.find('tokenizers/punkt_tab/english/')
except LookupError:
    print("NLTK 'punkt_tab' tokenizer data not found. Downloading...")
    nltk.download('punkt_tab')
    print("Download complete.")

# Now that 'punkt' and 'punkt_tab' are guaranteed to be downloaded, perform tokenization
sent_tokenize(corpus)
```
