```python
positive_words = ["happy", "good", "great", "awesome", "love"]
negative_words = ["sad", "bad", "terrible", "hate", "awful"]

text = "I hate the way do coding Aryan, Keep it up, your work in the field of coding is really great and i feel bad for you keep it up aryan but u can do good!"
# their are 3 positive words so i think that it will give me a positive feed back lets try it 
# now lets try for negative feedback ok 
# now lets try to bring out the neutral feedback from it by providing the equal no of positive and negative feed back ok
positive_count = sum(text.lower().count(word) for word in positive_words)
negative_count = sum(text.lower().count(word) for word in negative_words)

if positive_count > negative_count:
    print("Positive sentiment ðŸ˜ŠðŸ‘")
elif negative_count > positive_count:
    print("Negative sentiment ðŸ˜¢ðŸ’”")

else: 
    print("Neutral sentiment ðŸ˜‰ðŸ’›")

#Thanks so this is what sentement analysis is all about .
    
```

```python
#Using TextBlob (Beginner-Friendly NLP Library)
from textblob import TextBlob

text = "The movie was good, but the ending was disappointing."

analysis = TextBlob(text)

print("Polarity:", analysis.sentiment.polarity)
print("Subjectivity:", analysis.sentiment.subjectivity)

if analysis.sentiment.polarity > 0:
    print("Positive ðŸ˜‹")
elif analysis.sentiment.polarity < 0:
    print("Negative ðŸ˜ ")
else:
    print("Neutral ðŸ˜")
    
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

text = "This phone is ok! The battery dies so fast ðŸ˜¡"

scores = analyzer.polarity_scores(text)
print(scores)

if scores['compound'] >= 0.05:
    print("Positive ðŸ˜Š")
elif scores['compound'] <= -0.05:
    print("Negative ðŸ˜ ")
else:
    print("Neutral ðŸ˜")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import pandas as pd

data = {
    "text": [
        "I love this product!",
        "This is terrible.",
        "Great experience.",
        "Worst service ever.",
        "It's okay."
    ],
    "label": ["positive", "negative", "positive", "negative", "neutral"]
}
df = pd.DataFrame(data)
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df["text"])

model = MultinomialNB()
model.fit(X, df["label"])

test_text = ["This is amazing"]
test_vector = vectorizer.transform(test_text)
prediction = model.predict(test_vector)

print("Prediction:", prediction[0])
```
