```python
text = "Tokenization is the first step in NLP."
tokens = text.split()
print(tokens)
```

```python
import re 
text = "Don't hesitate to ask questions!"
tokens = re.findall(r"\w+|\S", text)
print(tokens)
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize

# Word tokenization
text = "This is a sample sentence. Here's another!"
words = word_tokenize(text)
print(words)
# Sentence tokenization
sentences = sent_tokenize(text)
print(sentences)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp("Tokenization is important for NLP.")
tokens = [token.text for token in doc]
print(tokens)
```
