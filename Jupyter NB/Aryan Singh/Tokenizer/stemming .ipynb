```python
import nltk
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer

nltk.download('punkt')

porter = PorterStemmer()
snowball = SnowballStemmer("english")
lancaster = LancasterStemmer()

words = ["running", "happiness", "flies", "fairly"]

print("PorterStemmer:",[porter.stem(word) for word in words])
print("SnowballStemmer:",[snowball.stem(word) for word in words])
print("LancasterStemmer:",[lancaster.stem(word) for word in words])
```
