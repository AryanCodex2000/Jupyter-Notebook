```python
tweet = "OMG!!! This is sooo coool #NerdsUnite ðŸ¤“ðŸ¤“. btw, u shud try it!"
```

```python
import re
def clean_text(text):
    text = re.sub(r'(\w)\1+', r'\1', text) 
    text = re.sub(r'[^\w\s#@]', '', text)  
    return text

cleaned = clean_text(tweet)
print(cleaned)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

doc = nlp(cleaned)
tokens = [token.text for token in doc]
print(tokens)
```
