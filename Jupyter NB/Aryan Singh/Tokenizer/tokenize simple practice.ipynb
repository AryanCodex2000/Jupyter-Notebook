```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize

text = "An ecosystem consists of all the living organisms (biotic factors) in a particular area, along with the non-living components (abiotic factors) of their environment, functioning together as a system. These components are interconnected through nutrient cycles and energy flows."

words = word_tokenize(text)
print("(This will show the words of the sentences one by one and each word will be seperated from the sentence ) words:", words)

sent = sent_tokenize(text)
print("(This will print hte sentence if we provided the paragraph then it will print in sentences): ", sent)
```
