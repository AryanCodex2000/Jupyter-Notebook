```python
from sklearn.feature_extraction.text import CountVectorizer

sentences = ["The cat sat on the mat"]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(sentences)
print("Vocab:", vectorizer.get_feature_names_out())
print("BoW:", X.toarray())
```
